```python
import sys
import tweepy
import json
import csv
import datetime
import time
import os
import re

c = 0
Trackers = ["$EURUSD", "$USDJPY", "$GBPUSD", "$USDCAD", "$USDCHF", "$AUDUSD"]
fileName =  '/Users/mitalipalsikar/Documents/year3project/Daily Twitter/' + datetime.datetime.now().strftime("%Y%m%d") + "twitterData.csv"
print("Writing to file ", fileName)

class StreamListener(tweepy.StreamListener):    
    
    '''
        def __init__(self, time_limit=60):
 
        self.time = time.time()
        self.limit = time_limit
    '''

        
    def on_connect(self):
        print("You are now connected to the streaming API.")
        global c
        c = 0
 
    def on_disconnect(self):
        print("You are now disconnected to the streaming API. ")
        
    def on_error(self, status_code):
        print('An Error has occured: ' + repr(status_code))
        return False
 
    def on_data(self, rdata):
        global c
        try:
            processTweet(rdata)
            c += 1
            print("Tweet No: ", c)
        except Exception as e:
            print(e)
            return False
        
#authenticate api 
def initAuth():
    ctoken = 'nmtuB46Ws1IO8JgLQr2BBmQi2'
    csecret = 'd0kZSPxnkFtaqwA1SsrOMsfcAr6OiHaVqmzWDNqrJ4ZOJvihpE'
    atoken = '1249023278-dReR46Ofl8SrgteF5gVKYNB2NoPDzSYiXKnYyJM'
    asecret = 'N7dddIfjUybmKXVfYnGxTmhKeqXiJdWCduIBwyWucOvXW'

    auth = tweepy.OAuthHandler(ctoken, csecret)

    auth.set_access_token(atoken, asecret)
    api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True, retry_count=10, retry_delay=5, retry_errors=5)
    return api

#start Twitter streaming
def startStream(api):
    start = time.time()
    while  (time.time() - start < 36000):
        try:
            listener = StreamListener(api=tweepy.API(wait_on_rate_limit=True)) 
            streamer = tweepy.Stream(auth=api.auth, listener=listener)
            print("Tracking: " + str(Trackers))
            streamer.filter(track=Trackers, languages=["en"], async=True)
        except:
            continue
    print("Finished execution at ", time.time())
    
#preproces text before saving into file
def processTweet(rdata):
    
    data = json.loads(rdata)
    tweet = data['text']
    tweet = ' '.join(re.sub("(@[A-Za-z0-9]+)|(\w+:\/\/\S+)"," ",tweet).split())

    tweetTime = data['timestamp_ms']
    date = datetime.datetime.fromtimestamp(int(tweetTime) / 1000)
    userName = data['user']['name']
    global fileName
    with open(fileName, 'a') as csvFile:
        csvWriter = csv.writer(csvFile)
        csvWriter.writerow([date,tweet, userName])

api = initAuth()
startStream(api)

with open(fileName, 'w') as csvFile:
    csvWriter = csv.writer(csvFile)
    csvWriter.writerow(['Date', 'Text', 'User'])
```
